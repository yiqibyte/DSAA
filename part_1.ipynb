```python
import time
import timeit
import matplotlib.pyplot as plt
```

```python
# 定义一个 O(n)时间复杂度的函数
# T(n) = n
def for_loop1(n: int) -> int:
    """for 循环"""
    res = 0
    # 循环求和 1, 2, ..., n-1, n
    for i in range(1, n + 1):
        res += i
    return res


# 绘制时间图
times = [timeit.timeit(stmt=f"for_loop1({i})", globals=globals()) for i in range(10, 101, 10)]
plt.plot(range(10, 101, 10), times)
plt.show()
```
